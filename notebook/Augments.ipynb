# Checking Augmentations

```python
data_path = os.path.join(os.getcwd(),'..','data')
augments_path = os.path.join(data_path, 'Augments')
image_names = [x for x in os.listdir(augments_path) if x.split('.')[-1] == 'png']
image_names.sort(key = lambda x: int(x.split('_')[-2].split('.')[0]))
```

```python
print('Total number of images {}'.format(len(image_names)))
for i in image_names:
    print(i.split('.')[0])
    print("=====")
```

```python
temp = []
for i in image_names:
    splitter = i.split('.')[0].split('_')
    if splitter[-2] + '_'+ splitter[-1] == 'Wound_1':
        temp.append(i)
print(len(temp))
temp
```

# Renaming Files

```python
for image_name in image_names:
    if image_name.split('_')[1] == 'groundtruth':
        new_name = 'SegMask_Wound_Bg_' +  image_name.split('_')[-3] +'_' + image_name.split('_')[-2].split('.')[0]+ '_'+ image_name.split('_')[-1]
        os.rename(os.path.join(augments_path, image_name), os.path.join(augments_path, 'SegMask_Wound_Bg',new_name))
    else:
        splitter = image_name.split('_')
        new_name = 'Original_' + splitter[-3] + '_' + splitter[-2].split('.')[0] + '_' + splitter[-1]
        os.rename(os.path.join(augments_path, image_name), os.path.join(augments_path, 'Original',new_name))
        print(new_name)
```
